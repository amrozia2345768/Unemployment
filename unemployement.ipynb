```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2: Load the dataset

```python
df = pd.read_excel("Pakistan_Poverty_Dataset_2000_2023.xlsx")
```

```python
print(df.head())
```

# Check the Missing values

```python
print(df.isnull().sum())
```

# Rename long column name

```python
df.rename(columns={
    'Unemployment Rate (%)': 'Unemployment_Rate'
}, inplace=True)

print(df[['Year', 'Unemployment_Rate']].head())
```

# Plot Unemployment Rate Over Time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Unemployment_Rate', data=df, marker='o', color='blue')
plt.title('Unemployment Rate in Pakistan (20002023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Compare with Other Economic Factors

```python
# Unemployment vs GDP Growth
plt.figure(figsize=(10, 5))
sns.scatterplot(x='GDP Growth Rate (%)', y='Unemployment_Rate', data=df)
plt.title('Unemployment Rate vs GDP Growth Rate')
plt.xlabel('GDP Growth Rate (%)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Unemployment vs Inflation Rate
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Inflation Rate (%)', y='Unemployment_Rate', data=df)
plt.title('Unemployment Rate vs Inflation Rate')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()

```

# Correlation Heatmap


```python
plt.figure(figsize=(12, 8))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap Between Economic Indicators')
plt.show()
```
